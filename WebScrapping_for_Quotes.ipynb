```python
import requests # for making standard html requests
from bs4 import BeautifulSoup # magical tool for parsing html data
import json # for parsing data
from pandas import DataFrame as df # premier library for data organization
import pandas as pd
```

```python
page = requests.get("https://www.goodreads.com/quotes/tag/technology")
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
```

```python
# Finding Quotes Div
quotes_div = soup.find_all(class_ = 'quoteText')
quotes = []
for item in quotes_div:
    script = item.select("script")
    for s in script:
        s.extract()
    raw_text = item.get_text()
    quote = raw_text[:raw_text.index('  ―')]
    author = item.find(class_ = 'authorOrTitle').text
    if len(quote) < 140:
        quotes.append({'quote': quote.strip().replace("“",""), 'author': author.strip().replace("“","")})


print(quotes)
quotes_df = df(quotes)
print(quotes_df)
```
