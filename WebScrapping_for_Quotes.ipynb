```python
import requests # for making standard html requests
from bs4 import BeautifulSoup # magical tool for parsing html data
import json # for parsing data
from pandas import DataFrame as df # premier library for data organization
import pandas as pd
```

```python
quotes = []
for i in range(1,100):
    page = requests.get("https://www.goodreads.com/quotes/tag/inspirational?page={}".format(i))
    soup = BeautifulSoup(page.text, 'html.parser')
    # Finding Quotes Div
    quotes_div = soup.find_all(class_ = 'quoteText')
    if quotes_div:
        for item in quotes_div:
            script = item.select("script")
            for s in script:
                s.extract()
            raw_text = item.get_text()
            quote = raw_text[:raw_text.index('  ―')]
            author = item.find(class_ = 'authorOrTitle').text
            if len(quote) < 140:
                quotes.append({'quote': quote.strip().replace("“","").replace("”", ""), 'author': author.strip().replace("“","")})


# Writing to CSV
print(quotes)
#quotes_df = df(quotes)
#df.to_csv('quotes.csv')
```

```python
import pandas as pd

def clean_author_data(val):
    if val.endswith(","):
        return val[:-1]
    return val

frame = pd.DataFrame(quotes)
print(frame.head())
frame['author'] = frame['author'].apply(lambda x: clean_author_data(x))
frame.to_csv('quotes.csv')
```
